```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("cevresel-veriler.xlsx", dtype=float)
data.head()
```

```python
# cevap vektörümüz
y = np.array(data["PM2.5"])
y
```

```python
# regresör matrisimizi oluşturalım
x1 = data.drop("PM2.5", axis=1)
x1.head()
```

```python
# regresör matrisimize B0 ı temsil eden birlerden oluşan sütunumuzu ekleyelim
x1["B0"] = np.ones((194,1), dtype=int)
x1.head()
```

```python
# B0 sütunumuzu en başa çekelim
col = ["B0","SO2","NO2","NOX","O3"]
x = x1.loc[:,col]
x
```

```python
# Betaşapka = [x'x]tersi * x'y
transpose_x = np.transpose(x)
transpose_x
```

```python
# tersini alalım
tersi_alınacak_degisken = transpose_x.dot(x)
tersi_alınacak_degisken
```

```python
inverse_x = np.linalg.inv(tersi_alınacak_degisken)
inverse_x
```

```python
# x'y yi bulalım
c = transpose_x.dot(y)
c
```

```python
#beta sapkayı bulalım
beta_sapka = inverse_x.dot(c)
beta_sapka
```
